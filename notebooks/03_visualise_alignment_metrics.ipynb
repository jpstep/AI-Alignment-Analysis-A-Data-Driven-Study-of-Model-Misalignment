# Alignment Metrics Visualisation

This notebook visualises aggregated alignment metrics derived from model responses.

Only summary statistics are used; no raw model outputs are loaded.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/summary_metrics.csv")
df
```

```python
metrics = df[df['metric'].isin(['refusal_rate', 'hedging_rate'])]

plt.bar(metrics['metric'], metrics['value'])
plt.title('Alignment Behaviour Rates')
plt.ylabel('Rate')
plt.show()
```
